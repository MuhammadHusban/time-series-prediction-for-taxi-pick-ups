```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Load the data
data = pd.read_csv('/content/drive/MyDrive/manhattan_final_prep_all_months.csv')
data = data[['tpep_pickup_datetime', 'PU_count']]

# Aggregate the data by hour
data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])
data = data.set_index('tpep_pickup_datetime')
data = data.resample('H').sum()

# Split the data into training and testing sets
train_size = int(len(data) * 0.8)
train_data, test_data = data[:train_size], data[train_size:]

# Scale the data
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data)
test_scaled = scaler.transform(test_data)

# Split the data into input (X) and output (y) variables
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X[i:i + time_steps, :]
        Xs.append(v)
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 24
X_train, y_train = create_dataset(train_scaled, train_scaled[:, 0], time_steps)
X_test, y_test = create_dataset(test_scaled, test_scaled[:, 0], time_steps)

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the LSTM model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, shuffle=False)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Inverse scale the predictions and the actual values
y_pred_inv = scaler.inverse_transform(np.concatenate((y_pred, X_test[:, -1, 1:]), axis=1))[:, 0]
y_test_inv = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), X_test[:, -1, 1:]), axis=1))[:, 0]

# Create a dataframe of the actual and predicted values
results = pd.DataFrame({'Actual': y_test_inv, 'Predicted': y_pred_inv}, index=test_data.index[time_steps:])

# Visualize the results
plt.figure(figsize=(15,5))
plt.plot(results.index, results['Actual'], label='Actual')
plt.plot(results.index, results['Predicted'], label='Predicted')
plt.xlabel('Date')
plt.ylabel('Pickup Count')
plt.title('Actual vs. Predicted Pickup Count by Hour')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import datetime

# Generate 24 hours of time steps for prediction
next_24_hours = np.arange(1, 25)

# Get the predicted pickups for the next 24 hours
predicted_pickups = model.predict(X_test)[-24:]

# Get the current time
now = datetime.datetime.now()

# Generate a list of datetime objects for the next 24 hours
next_24_times = [now + datetime.timedelta(hours=i) for i in range(1, 25)]

# Plot the predicted pickups
plt.figure(figsize=(15,5))
plt.plot(next_24_times, predicted_pickups, label='Predicted Pickups')

# Add labels and title to the plot
plt.xlabel('Time')
plt.ylabel('Number of Pickups')
plt.title('Predicted Pickups for the Next 24 Hours')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the legend and the plot
plt.legend()
plt.show()

# Plot the loss function over time
plt.figure(figsize=(15,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Time')
plt.legend()
plt.show()

# Plot the actual vs. predicted values over time
plt.figure(figsize=(15,5))
plt.plot(results.index, results['Actual'], label='Actual')
plt.plot(results.index, results['Predicted'], label='Predicted')
plt.xlabel('Date')
plt.ylabel('Pickup Count')
plt.title('Actual vs. Predicted Pickup Count by Hour')
plt.legend()
plt.show()

# Plot the distribution of actual and predicted values
plt.figure(figsize=(15,5))
plt.hist(results['Actual'], bins=50, alpha=0.5, label='Actual')
plt.hist(results['Predicted'], bins=50, alpha=0.5, label='Predicted')
plt.xlabel('Pickup Count')
plt.ylabel('Frequency')
plt.title('Distribution of Actual and Predicted Pickup Counts')
plt.legend()
plt.show()
```
